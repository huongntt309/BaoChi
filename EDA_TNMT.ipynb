## 9. Plot different features against one another

### Histogram


```
# Thay thế các nhãn trùng lặp trong cột 'Linh_vuc'
df['Linh_vuc'].replace('KTTV -BĐKH', 'KTTV-BĐKH', inplace=True)
df['Linh_vuc'].replace('KTTV - BĐKH', 'KTTV-BĐKH', inplace=True)

# Tạo subplot để vẽ cả bar chart và pie chart
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

# Vẽ bar chart
df['Linh_vuc'].value_counts().plot(kind='bar', color='skyblue', ax=axes[0])
axes[0].set_title("Phân bố Lĩnh vực (Bar chart)")
axes[0].set_ylabel('Số bài viết')
axes[0].set_xlabel('Lĩnh vực')

# Vẽ pie chart
df['Linh_vuc'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[1])
axes[1].set_title("Phân bố Lĩnh vực (Pie chart)")
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```
import re

# Sử dụng biểu thức chính quy để tìm và thay thế dấu cách thừa trong 'Chuyen_muc'
df['Chuyen_muc'] = df['Chuyen_muc'].apply(lambda x: re.sub(r'\s+', ' ', x.strip()) if 'thông tin khác' in x else x)

# Tạo subplot để vẽ cả bar chart và pie chart
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 7))

# Vẽ bar chart
df['Chuyen_muc'].value_counts().plot(kind='bar', color='lightgreen', ax=axes[0])
axes[0].set_title("Phân bố Chuyên mục (Bar chart)")
axes[0].set_ylabel('Số bài viết')
axes[0].set_xlabel('Chuyên mục')

# Vẽ pie chart
df['Chuyen_muc'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[1])
axes[1].set_title("Phân bố Chuyên mục (Pie chart)")
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

## EDA Sum

```
# Tính độ dài của mỗi đoạn tóm tắt
df['Tom_tat_noi_dung_length'] = df['Tom_tat_noi_dung'].apply(lambda x: len(x.split()))

# Phân tích từ vựng
unique_words = set()
for summary in df['Tom_tat_noi_dung']:
    unique_words.update(summary.split())

# Tính toán số lượng từ vựng và độ dài trung bình
vocab_size = len(unique_words)
avg_summary_length = df['Tom_tat_noi_dung_length'].mean()

# Hiển thị kết quả
print("Số lượng từ vựng trong các đoạn tóm tắt (chia theo dấu cách):", vocab_size)
print("Độ dài trung bình của các đoạn tóm tắt (chia theo dấu cách):", avg_summary_length)

# Vẽ histogram cho độ dài của các đoạn tóm tắt
plt.hist(df['Tom_tat_noi_dung_length'], bins=30, color='skyblue', edgecolor='black')
plt.title("Phân phối độ dài của các đoạn tóm tắt")
plt.xlabel("Độ dài")
plt.ylabel("Số lượng")
plt.show()
```
