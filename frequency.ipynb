```python
from tqdm import tqdm
```

```python
count_all = dict()
for t in df.golden_subtopic.unique():
  count_all[t] = Counter()
for i, r in tqdm(df.iterrows()):
  count_all[r.golden_subtopic] |= Counter(word_tokenize(re.sub(r'[^\w\s]','',r.Noi_dung_tin_bai.lower())) )
```

```python
for d in count_all:
  count_all[d] = count_all[d].most_common()
```

```python
for d in count_all:
  count_all[d] = [p for p in count_all[d] if len(p[0].split()) > 1]
```

```python
count_all['Đất đai']
```

```python
list_of_df = list()
for d in count_all:
  tdf = pd.DataFrame.from_dict(count_all[d][:200])
  tdf.columns = [d, '{}_count'.format(d)]
  list_of_df.append(tdf)
```

```python
df_all = pd.DataFrame()
```

```python
for dd in list_of_df:
  df_all.join(dd)
```

```python
pd.concat(list_of_df, axis = 1).to_excel('count_all_label.xlsx')
```
